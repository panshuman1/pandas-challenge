```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_school= school_data["School ID"].count()
print(f"Total number of schools :{total_school}")

total_student= student_data["Student ID"].count()
print(f"Total number of students :{total_student}")

total_budget= school_data["budget"].sum()
print(f"Total Budget :{total_budget}")

avg_math_score= student_data["maths_score"].mean()
avg_math_score= round(avg_math_score,6)
print(f"Average math score :{avg_math_score}")


avg_reading_score= student_data["reading_score"].mean()
avg_reading_score= round(avg_reading_score,6)
print(f"Average reading score :{avg_reading_score}")


pass_count_s= student_data[student_data['maths_score'] >=50].count()
pass_count=pass_count_s["Student ID"]
passing_percentage_math = round(pass_count/total_student*100,6)
print(f"Passing percentage math :{passing_percentage_math}")

pass_count_s1= student_data[student_data['reading_score'] >=50].count()
pass_count_r=pass_count_s1["Student ID"]
passing_percentage_read = round(pass_count_r/total_student*100,6)
print(f"Passing percentage read :{passing_percentage_read}")

#Calculating over all_passing_percentage
df= student_data[(student_data['maths_score'] >=50) & (student_data['reading_score'] >=50)]
total_passingcount=df["Student ID"].count()
overall_passing_per= round(total_passingcount/total_student*100,6)
print(f"Over All Passing Percentage :{overall_passing_per}")


####################################################################################################
area_summary  =pd.DataFrame({
    'Total Schools': [total_school],
    'Total Students': [total_student],
    'Total Budget': [total_budget],
    'Average Maths Score':[avg_math_score],
    'Average Reading Score':[avg_reading_score],
    '% Passing Maths':[passing_percentage_math],
    '% Passing Reading':[passing_percentage_read],
    '% Overall Passing':[overall_passing_per]
    })

area_summary 
####################################################################################################
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_summary_df10
```

```python
school_spending_df= school_summary_df10
#This command can run only once
#del school_spending_df["School Type"]

school_spending_df["Spending Per Student"]=school_spending_df["budget"]/school_spending_df["Total Students"]

#Added this line as the pd.cut was giving error ValueError: setting an array element with a sequence
spending_perstud= school_spending_df["budget"]/school_spending_df["Total Students"]
school_spending_df

```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
school_spending_df["Spending Ranges(Per Student)"]=pd.cut(np.array(school_spending_df["Spending Per Student"]),spending_bins,labels=labels)
school_spending_df
```

```python
spending_maths_scores = school_spending_df.groupby(["Spending Ranges(Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges(Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = school_spending_df.groupby(["Spending Ranges(Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges(Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges(Per Student)"]).mean()["% overall passing"]
type(spending_maths_scores)
```

```python
df_spending_maths_scores= pd.DataFrame(spending_maths_scores)
df_spending_reading_score = pd.DataFrame(spending_reading_scores)
df_spending_passing_maths = pd.DataFrame(spending_passing_maths)
df_spending_passing_reading = pd.DataFrame(spending_passing_reading)
df_overall_passing_spending = pd.DataFrame(overall_passing_spending)

spending_summary   = pd.concat([df_spending_maths_scores, df_spending_reading_score, df_spending_passing_maths,df_spending_passing_reading,df_overall_passing_spending ]
                                         , axis=1)
spending_summary  
```
